# Updated App

```python
menu = {'coffee': 80, 'salad': 50, 'taco': 30, 'cold drink': 40, 'sandwich' : 20, 'tea': 10 }
menu.keys()
menu.values()
'COFEE'.lower() in menu.keys()
print(menu)
n = input("Enter name" '\n')

print("Welcome to our cafe!\n")
print("This is our menu for today. Please order and enjoy your dishes\n")
print(menu)

orders = []
total_amount = 0

while True:
  user_input = input("please select your item from the menu or enter q/quite to exit\n")

  if user_input.lower() in menu.keys():
    total_amount = total_amount + menu[user_input.lower()]
    orders.append(user_input)
    print(f"You have ordered {orders} and the total amount of bill is ${total_amount}\n")
  elif user_input.lower() == 'q' or user_input.lower() == 'quite':
    print("Thank you for your order.\n it will be served you soon\n")
    tip_bool = input("Do you want to add a tip for the staff? (Yes/No\n)")
    if tip_bool.lower() == "yes":
      tip_amount = int(input("Enter the tip amount\n"))
      tip_amount = total_amount + tip_amount
      print(f"you have orders {orders} and the total amount of bill is ${tip_amount}\n")
    elif tip_bool.lower() == "no":
      print(f"you have orders {orders} and the total amount of bill is ${total_amount}\n")
    break
  elif user_input.lower() not in menu.keys():
    print(f"you have ordered {orders} and  the total amount of bill is ${total_amount}\n")
  else:
    print("please enter a valid input\n")
    break
```
