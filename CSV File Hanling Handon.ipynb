```python
def csv_file (file_path, search_field, search_term):
    results = []
    with open (file_path , 'r') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader)

        if search_field not in header:
            print(f"Error: '{search_field}' is not a valid search field.")
            return results
        for row in csv_reader:
          if search_term.lower() in row[header.index(search_field)].lower():
            results.append(row)

    return results


file_path = '/content/example_transactions (2).csv'
while True:
  search = input("Search by Name, Age, Department or Salary or else enter quit to Terminate \n ")
  search_field = search.capitalize()
  if search.lower() == 'quit':
    print("User wants to Terminate the Program. See you later!\n")
    break
  elif search_field not in ['Name', 'Age', 'Department', 'Salary']:
    print("ERROR: No such column exists in the given CSV file. You must enter a valid search field.\n")
  else:
    value = input("What do you want to search for?\n ")
    results = csv_file(file_path, search_field, value)

    if results:
        print("\n".join([", ".join(row) for row in results]))
        print(f"\n{len(results)} records found.")
    else:
        print("\n 0 records found.")
```

```python
import csv

# Function to search CSV
def search_csv(file_path, search_field, search_term):
    results = []
    with open(file_path, 'r') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader)

        if search_field not in header:
            print(f"ERROR: '{search_field}' is not a valid search field.")
            return results

        for row in csv_reader:
            if search_term.lower() in row[header.index(search_field)].lower():
                results.append(row)

    return results

# Read header once at the start
file_path = "/content/example_transactions (2).csv"

# Read the header dynamically
with open(file_path, 'r') as file:
    csv_reader = csv.reader(file)
    header = next(csv_reader)

while True:
    search = input(f"Search by {', '.join(header)} or enter 'quit' to Terminate:\n ")
    search_field = search.capitalize()

    if search.lower() == 'quit':
        print("User wants to Terminate the Program. See you later!\n")
        break

    elif search_field not in header:
        print("ERROR: No such column exists in the given CSV file. You must enter a valid search field.\n")
    else:
        value = input("What do you want to search for?\n ")
        results = search_csv(file_path, search_field, value)
        if results:
            print("\n".join([", ".join(row) for row in results]))
            print(f"\n{len(results)} records found.")
        else:
            print("\n0 records found.")
```
