```python
import csv
```

```python
def Search_csv (file_path,search_field,Value):
  results =[]

  with open(file_path,'r') as file:
    file_read = csv.reader(file)
    header = next(file_read)

    if search_field not in header:
      print(f"Error; '{search_field} not a valid search.")
      return results

    for row in file_read:
      if Value.lower() in row[header.index(search_field)].lower():
        results.append(row)

  return results

file_path = "/content/example_transactions.csv"

while True:
  Search = input("Search by name,Age,Department or salary or else quit to terminate the program\n")
  search_field = Search.capitalize()

  if Search.lower() == 'quit':
    print("You want to terminate,See you later!")
    break
  elif search_field not in [ 'Name','Age','Department','Salary']:
    print("Error: NO such coloumn Exists,Enter a valid field\n")
  else:
    Set = input("What do you want to search for?\n")
    results = Search_csv(file_path,search_field,Set)

    if results:
      print("\n".join([",". join(row) for row in results]))
      print(f"\n {len(results)}) record found.")
    else:
      print("0 Record found")
```
