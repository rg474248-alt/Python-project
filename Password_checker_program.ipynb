```python
# Function to evaluate password strength
#def check_password(password):

# Check for Excellent password quality
# Must be longer than 15 characters and contain:
# uppercase, lowercase, digit, and special character

# Main program loop
#while True:

# Allow user to exit the application
#if password.lower() == 'quit':




def check_password(password):
  if len(password) > 15 and any(c.isupper() for c in password ) and any (c.islower() for c in password) and any (c.isdigit() for c in password) and any (not c.isalnum() for c in password):
    return "Excellent"
  elif 12<= len(password) <=14 and any(c.isupper()  for c in password) and any (c.islower() for c in password) and any (c.isdigit() for c in password) and any (not c.isalnum() for c in password):
    return "Strong"
  elif 9<= len(password) <=11 and  any(c.isupper() for c in password )and any (c.islower() for c in password) and any (c.isdigit() for c in password) and any (not c.isalnum() for c in password):
    return "Medium"
  elif 6<= len(password) <=8  and  any(c.isupper() for c  in password) and any (c.islower() for c in password) and any (c.isdigit() for c in password) and any (not c.isalnum() for c in password):
    return "Weak"
  else:
    return "Unsafe"

print("Password Quality Program:\n")

while True:

  password = input("enter your password, or 'quit/Quit' to exit: ")
  print("password:", password)

  if password.lower() == 'quit':
    break

  output = check_password(password)
  print('Your password is considered:', output)

```
